````python
# function to get story specefic analytics (24h refresh)
def get_analytics_story(stories_df):
  views=[]
  revenue=[]
  viewers=[]
  paidTwistUnlockers=[]
  freeTwistUnlockers=[]
  freeConversion=[]
  paidConversion=[]
  all_params = ["views", "revenue", "viewers", "paidTwistUnlockers", "freeTwistUnlockers", "freeConversion", "paidConversion" ]
  columns_dict = {
                  'views': views,
                  'revenue': revenue,
                  'viewers': viewers,
                  'paidTwistUnlockers': paidTwistUnlockers,
                  'freeTwistUnlockers': freeTwistUnlockers,
                  'freeConversion': freeConversion,
                  'paidConversion': paidConversion
                    }
  for i, row in stories_df.iterrows():
      headers = {
        'x-auth-token': 'OTd5eE5KNFFlT0trNjRPMzoxZDdjYTJiZTFhZDZmMzQ4YjNlZDQ2OGRlNTU5M2JjMTBlNmU3YTA0M2FhYjYyMWE1MDM3NmVlMDhlMWZiZjZi',
        'Content-Type': 'application/json'
      }
      url = f"https://story3.com/api/v2/analytics/{row['hashId']}"
      response = requests.request("GET", url, headers=headers)
      json_res = json.loads(response.text)
      print(json_res)
      res_keys = list(json_res.keys()) # list of api reponse keys
      # appending permenant and conditonal api keys to their respective fields
      for key in all_params:
        if key not in res_keys:
          columns_dict[key].append(0)
        else :
          columns_dict[key].append(json_res[key])

  return columns_dict
columns_dict = get_analytics_story(stories_df)
stories_df = stories_df.assign(**columns_dict)
````
